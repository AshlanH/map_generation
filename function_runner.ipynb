### PLOT FUNCTION - DYNAMIC ADJUSTMENT BASED ON NEIGHBORING VALUES

```python
math.exp(-3 * 0.8)
```

```python
def calculate(x, neighbor_mean, alpha):
    return math.exp(-alpha * x) * (neighbor_mean - x)


f, ax = plt.subplots(1, 1, figsize = (16, 8))

def plot(ax, cur_val, alpha):
    neighbor_mean = np.arange(0, 1, 0.1)
    sns.lineplot(
        x = neighbor_mean,
        y = [calculate(cur_val, i, alpha) for i in neighbor_mean],
        ax = ax,
        label = 'x = {}'.format(round(cur_val, 3))
)

for each_value in np.arange(0, 1, 0.1):
    plot(ax, each_value, 5)

# ax.set_xticks(np.arange(0, 1, 0.1))
# ax.set_ybound(0.25, 1)
plt.xlabel("Mean Neighbor Value")
plt.ylabel("Increment")
plt.legend()
plt.grid()
plt.show()
```

### PLOT FUNCTION - EXPONENTIAL SCALE IN RELATION TO STATE

```python
def calculate(x, alpha):
    return math.exp(-alpha * x)


f, ax = plt.subplots(1, 1, figsize = (16, 8))

def plot(ax, alpha):
    origin_value = np.arange(0, 1, 0.1)
    sns.lineplot(
        x = origin_value,
        y = [calculate(i, alpha) for i in origin_value],
        ax = ax,
        label = 'Alpha = {}'.format(round(alpha, 3))
)

for alpha in [3, 3.5, 4, 4.5, 5]: ## EACH ALPHA VALUE
    plot(ax, alpha)

# ax.set_xticks([0.4 - i for i in np.arange(0, 1, 0.1)])
# ax.set_ybound(0.25, 1)
plt.xlabel("Origin Value")
plt.ylabel("Output")
plt.legend()
plt.grid()
plt.show()

# HIGHER ALPHA = LARGER CHANGES
```
